## KNN Imputation - Cylinders

```python
from sklearn.impute import KNNImputer
```

```python
car_dt['cylinders']
```

```python
car_dt['cylinders'] = car_dt['cylinders'].str.slice(stop = 2)
```

```python
car_dt['cylinders']
```

```python
car_dt['cylinders'].unique()
```

```python
#Converting "other" entries to "nan"
car_dt['cylinders'] = car_dt['cylinders'].replace(['ot'], [np.nan])
car_dt.cylinders = car_dt.cylinders.astype(float)
car_dt.cylinders.dtype
```

```python
car_dt['cylinders'].dtype
```

```python
car_dt['odometer'].dtype
```

```python
car_dt
```

```python
#KNN Imputer
imputer = KNNImputer(n_neighbors = 2)
```

```python
df_knntest = car_dt[['cylinders', 'odometer']]
```

```python
#df_filled = imputer.fit_transform(car_dt[['cylinders','odometer']])
df_filled = imputer.fit_transform(df_knntest)
```

```python
df_filled
```

```python
car_dt[['cylinders','odometer']] = df_filled
```
