```python
dimcols=['CensusTract', 'LAT', 'LONG', 'Index', 'Standardindex', 'CensusTract.1',
       ' Correlation_Index ', 'Address','Poverty_prob_index',
       'MHLTH', 'PHLTH']

Index_datadim=Index_data[dimcols]
```
